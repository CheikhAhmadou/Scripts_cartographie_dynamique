# Carte de chaleur

On est en présence d'un phénomène qui a une certaine intensité et se développe autour d'un point géographique

1. Faire une carte folium centrée en un point et le rendre visible avec un Marker
2. générer un jeu de données arbitraire et qui sera stocké dans un dataframe pour N point
3. convertir les tableaux en listes
4. Employer le plugin HeatMap pour rendre compte du phénomène.

```python
import os
import folium
import numpy as np
import pandas as pd
```

On se définit un nb de points à suivre Point p(x,y) et une intensité du phénomène représenté par une valeur réelle (liste 3)

```python
N = 1000 # le nb de points
# pour les coordonnées x, y
liste1 = [ 48.0 + np.random.normal(0,3) for i in range(N) ]
liste2 = [ 5.0 + np.random.normal(0,2) for i in range(N) ]
# pour l'intensité du phénomène
liste3 = [1.0 + np.random.exponential(2,) for i in range(N)]

# refaire la liste de Point mais en version POO
```

pour décrire le phénomène on dispose de données latitude longitude et intensité.
On commence par charger nos listes dans un DataFrame

```python
tableauDonnees = pd.DataFrame()
tableauDonnees['Lat'] = liste1
tableauDonnees['Long'] = liste2
tableauDonnees['intensité'] = liste3
```

```python
# conversion vers un tableau numpy
tab = tableauDonnees[['Lat', 'Long', 'intensité']].to_numpy()
```

```python
type(tab) # on est passé de 1 dataframe à un tableau numpy
```

```python
nos_data = tab.tolist()  # on passe d'un tableau numpu à une liste
```

```python
#type(nos_data) # pour vérifier
# nos_data        # la liste qui est attendue par le plugin
```

## une Carte Folium

```python
# objet Carte
maCarte = folium.Map(
                location = [48.3, 5.0],
                tiles = "openstreetmap",
                zoom_start = 6
            )
# objet Marker
marqueur = folium.Marker(location=[48.0,5.0])
marqueur.add_to(maCarte)
#maCarte.add_child(marqueur)
```

```python
# un troisième objet, l'objet HeatMap instancié
from folium.plugins import HeatMap

hm = HeatMap(nos_data)
hm.add_to(maCarte)
```
