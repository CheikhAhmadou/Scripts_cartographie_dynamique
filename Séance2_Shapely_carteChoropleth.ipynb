```python
# définissons un carré avec un trou
frontiere = [(-180,90), (-180,-90), (180, -90), (180,90)]
```

```python
# on utilise une liste de listes pour les trous (holes)
trou = [[(-170,50), (-170,-50), (120, -50), (120,50)], 
         [(-170,70), (-170,60), (120, 60), (120,70)] ]
```

```python
univ = Polygon(shell=frontiere, holes=trou)
```

```python
univ
```

 Pour un objet polygone on peut récupérer via des attributs ou méthodes des informations assez "naturelles"

```python
print(univ.centroid)
```

```python
print(univ.area) # surface
```

```python
print(univ.bounds) # la bounding box
```

```python
print(univ.exterior) # périmètre
```

```python
print(univ.exterior.length) # vaeur du périmètre
```

# 2ème partie : faire une carte Folium

```python
import os
import folium, json
from folium import plugins
```

```python
# on fait un fond de carte
maCarte = folium.Map(
                    location = [38.76, 107.24] ,
                    tiles = "cartodbpositron",
                    zoom_start = 8
                )
```

# import des géométries

```python
# importer des geometries, il faut donc disposer d'un fichier de geometries
# le fichier de données s'appelle par exemple
geom = os.path.join('data_divers' ,'china_geom.geojson' )
provincesDeChine = json.load(open(geom))
```

```python
print(type(geom))
```

```python
print(type(provincesDeChine))
```

# import des données

```python
import pandas
population2010 = os.path.join('data_divers','china_pop.csv')
popChine2010 = pandas.read_csv(population2010)
```

```python
popChine2010.head()
```

```python
df = pandas.DataFrame()
```

```python
df['Nom'] = popChine2010['name']
```

```python
df['Population'] = popChine2010['population2010']
```

```python
df.head()
```

## objet Choropleth

```python
c = folium.Choropleth(
    geo_data = provincesDeChine,
    data = df,
    columns = ['Nom', 'Population'],
    key_on= 'feature.properties.name',
    fill_color='YlOrBr',
    fill_opacity = 0.9,
    line_opacity = 0.2,
    legend_name = "Population par province",
    highlight=True,
    name = "Population Chine",
    bins=8
    )
```

```python
type(c)
```

```python
c.add_to(maCarte)
```

```python
maCarte
```

```python
lc = folium.LayerControl()
```

```python
lc.add_to(maCarte)
```
